```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(15,6))
x = np.linspace(0,10,50)
x_samps = np.linspace(0.5,9,11)

ax1.set(xticks=[], yticks=[], xlim=(0,10), ylim=(-4,4), title='Amplitude values without grid')
ax1.set_xlabel('Time', x=1)
ax1.set_ylabel('Amplitude')
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['bottom'].set_position('zero')
ax1.plot(x,np.sin(x)*2.95, c='b')
ax1.bar(x_samps,np.sin(x_samps)*2.95,width=0.05, color='k')
ax1.scatter(x_samps,np.sin(x_samps)*2.95, c='r')

ax2.set(xticks=[], yticks=[-3.5, -2.5, -1.5, -0.5, 0.5, 1.5, 2.5, 3.5], xlim=(0,10), ylim=(-4,4))
ax2.set_xlabel('Time', x=1)
ax2.set_ylabel('Amplitude in 3 bit resolution')
ax2.set_yticklabels(['000', '001', '010', '011', '100', '101', '110','111'])
ax2.set_title('Amplitude values with 3 bit grid')
ax2.grid()
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['bottom'].set_position('zero')
ax2.plot(x,np.sin(x)*2.95, c='b')
y_smps = np.round(np.sin(x_samps)*2.95+0.5)-0.5
ax2.bar(x_samps,y_smps,width=0.05, color='k')
ax2.scatter(x_samps,y_smps, c='r')

fig.savefig('01A_bitdepth.png', bbox_inches='tight')
```
